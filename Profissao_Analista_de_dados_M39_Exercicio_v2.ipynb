```
# Instale o boto3, caso ainda não esteja instalado
!pip install boto3

import boto3
import pandas as pd
from getpass import getpass
from botocore.exceptions import ClientError  # Import correto da exceção

# Solicita as credenciais AWS ao usuário
aws_access_key_id = getpass('Enter your AWS Access Key ID: ')
aws_secret_access_key = getpass('Enter your AWS Secret Access Key: ')

# Cria o cliente S3 usando as credenciais fornecidas
try:
    client = boto3.client(
        's3',
        aws_access_key_id=aws_access_key_id,
        aws_secret_access_key=aws_secret_access_key
    )
    print("✅ Conexão ao S3 criada com sucesso.")
except Exception as e:
    print(f"❌ Erro ao criar o cliente S3: {e}")
    exit()

# Define o bucket e os arquivos
BUCKET = 'modulo1-40-ebac-sagemaker'
KEY = 'cotacao (1).csv'  # Nome do objeto no S3
FILENAME = 'cotacao_aws.csv'  # Nome do arquivo a ser salvo localmente

# Verifica se o objeto existe no bucket
try:
    client.head_object(Bucket=BUCKET, Key=KEY)
    print(f"✅ O arquivo '{KEY}' existe no bucket '{BUCKET}'.")
except ClientError as e:
    if e.response['Error']['Code'] == '404':
        print(f"❌ Erro: Arquivo '{KEY}' não encontrado no bucket '{BUCKET}'.")
    else:
        print(f"❌ Erro ao verificar o arquivo: {e}")
    exit()  # Encerra o script caso o arquivo não exista

# Realiza o download do arquivo se ele existir
try:
    client.download_file(BUCKET, KEY, FILENAME)
    print(f"✅ Arquivo '{KEY}' baixado com sucesso como '{FILENAME}'.")
except ClientError as e:
    print(f"❌ Erro ao baixar o arquivo: {e}")
    exit()

# Lê o arquivo baixado e exibe os dados
try:
    df = pd.read_csv(FILENAME)
    print(f"✅ Arquivo CSV lido com sucesso. Primeiras linhas:\n{df.head()}")
except Exception as e:
    print(f"❌ Erro ao ler o arquivo CSV: {e}")
```
